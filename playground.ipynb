```python
def copy_file(file_id):
    ktp_file = os.path.join(KTP_DIR, f"{file_id}.jpg")
    selfie_file = os.path.join(SELFIE_DIR, f"{file_id}.jpg")

    if os.path.exists(ktp_file) and os.path.exists(selfie_file):
        print(f"Copying File ID: {file_id} - KTP: {ktp_file}, Selfie: {selfie_file}")
        ktp_name = ktp_file.split("\\")[-1].replace(".jpg", "_1.jpg")
        selfie_name = selfie_file.split("\\")[-1].replace(".jpg", "_2.jpg")
        # check if the file already exists in the destination directory
        if os.path.exists(ktp_name) or os.path.exists(selfie_name):
            print(f"File ID: {file_id} - KTP or Selfie file already exists in the destination directory. Skipping copy.")
            return
        # Add actual file copying logic here
        shutil.copy(ktp_file, ktp_name)
        print(f"Copied to {ktp_name}")
        shutil.copy(selfie_file, selfie_name)
        print(f"Copied to {selfie_name}")

    else:
        print(f"File ID: {file_id} - Missing KTP file.") if not os.path.exists(ktp_file) else print(f"File ID: {file_id} - Missing Selfie file.")
```

```python
real_file_id = [x.replace('_2.jpg', '.jpg') for x in os.listdir() if x.endswith('_2.jpg')]
real_file_id
```

```python
real_json = {}

for file in real_file_id:
    _id = "".join(file.split('.')[:-1])
    # print(f"Processing file: {file} with ID: {_id}")
    real_json[_id] = {
        "ktp": f"https://raw.githubusercontent.com/kevinadityai/test/refs/heads/main/{file.replace('.jpg', '_1.jpg')}",
        "selfie_1": f"https://raw.githubusercontent.com/kevinadityai/test/refs/heads/main/{file.replace('.jpg', '_2.jpg')}"
    }

    
    spoof_path = f"{'.'.join(file.split('.')[:-1])}_3.{file.split('.')[-1]}"
    print(f"Checking for spoof file: {spoof_path}")
    if os.path.exists(spoof_path):
        print(f"File {file} has a corresponding _3 file. Adding to JSON.")
        real_json[_id]["selfie_2"] = f"https://raw.githubusercontent.com/kevinadityai/test/refs/heads/main/{spoof_path}"

real_json
```

```python
import random
import copy

# 1. Extract existing selfie_1 and selfie_2 values
selfie_1_list = [val['selfie_1'] for val in real_json.values() if 'selfie_1' in val]
selfie_2_list = [val['selfie_2'] for val in real_json.values() if 'selfie_2' in val]

# 2. Shuffle the extracted lists among their respective groups
random.shuffle(selfie_1_list)
random.shuffle(selfie_2_list)

# 3. Create the new dict and reassign the values
shuffled_data_json = {}

for key, val in real_json.items():
    # Keep KTP as the constant identifier
    shuffled_data_json[f"{key}_shuffled"] = {'ktp': val['ktp']}
    
    # Pop from the shuffled lists if the source data had these keys
    if 'selfie_1' in val:
        shuffled_data_json[f"{key}_shuffled"]['selfie_1'] = selfie_1_list.pop()
        
    if 'selfie_2' in val:
        shuffled_data_json[f"{key}_shuffled"]['selfie_2'] = selfie_2_list.pop()

# View the result
shuffled_data_json
```

```python
ai_file_id = [f"{x.split('.')[0][:-1]}.{x.split('.')[1]}" for x in os.listdir() if x.split('.')[0].endswith('A')]
ai_file_id
```

```python
ai_result_json = {}

for file in ai_file_id:
    ai_result_json[file.split('.')[0]] = {
        "ktp": f"https://raw.githubusercontent.com/kevinadityai/test/refs/heads/main/{file.split('.')[0]}A.{file.split('.')[1]}",
        "selfie_1": f"https://raw.githubusercontent.com/kevinadityai/test/refs/heads/main/{file.split('.')[0]}A.{file.split('.')[1]}"
    }

ai_result_json
```

```python
import json
# real_json + ai_result_json
data_json = {**real_json, **shuffled_data_json, **ai_result_json}
with open('data.json', 'w') as f:
    json.dump(data_json, f, indent=4)
data_json
```
